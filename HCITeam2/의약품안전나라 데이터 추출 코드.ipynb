### 알약 앞뒤에 마크 "없는" 데이터 저장하기

```python
# 여기서 정규식으로 마크 들어간 행 다 없앰. 
import urllib.request as req
import re
# https://jhproject.tistory.com/166

regexp = re.compile(r"마크")

nonMarkIndex=[]
for i in range(data.shape[0]):
  if (regexp.search(str(data.loc[i]["표시앞"])) == None) and (regexp.search(str(data.loc[i]["표시뒤"])) == None) :
    nonMarkIndex.append(i)
```

```python
nonMark = data.loc[nonMarkIndex]
```

### 분할선 있는 알약 데이터 저장하기 

```python
# 여기서 정규식으로 마크 들어간 행 다 없앰. 
import urllib.request as req
import re
# https://jhproject.tistory.com/166

regexp = re.compile(r"분할선")

partingLineIndex=[]
for i in range(data.shape[0]):
  if regexp.search(str(data.loc[i]["표시앞"])) or regexp.search(str(data.loc[i]["표시뒤"])) :
    partingLineIndex.append(i)
```

```python
partingLine = data.loc[partingLineIndex]
```
